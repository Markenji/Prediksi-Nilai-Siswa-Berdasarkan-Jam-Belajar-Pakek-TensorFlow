<h1>Prediksi Nilai Berdasarkan Jam Belajar Menggunakan Single Linear Regression</h1>

<h2>Pendahuluan</h2>
<p>Single Linear Regression adalah salah satu algoritma Machine Learning yang digunakan untuk memprediksi nilai variabel dependen (target) berdasarkan satu variabel independen (prediktor). Dalam konteks ini, kita akan menggunakan Single Linear Regression untuk memprediksi nilai siswa berdasarkan jumlah jam belajar mereka.  Meskipun sederhana, model ini bisa menjadi titik awal untuk memahami hubungan antara variabel dan membangun model yang lebih kompleks.</p>

<h2>Penerapan NLP (Natural Language Processing)</h2>
<p>Meskipun judulnya menyebutkan NLP, penerapan NLP secara langsung pada model regresi linear sederhana ini mungkin terbatas.  NLP biasanya digunakan untuk memproses teks dan mengekstrak informasi bermakna.  Dalam konteks prediksi nilai siswa berdasarkan jam belajar, NLP bisa diterapkan jika data jam belajar diambil dari teks, misalnya dari catatan harian siswa yang ditulis bebas.  Dalam kasus tersebut, kita akan menggunakan teknik NLP untuk mengekstraksi jumlah jam belajar dari teks tersebut sebelum memasukkannya ke model regresi linear. </p>
<p>Contohnya, jika catatan harian siswa mengandung kalimat "Saya belajar selama 2 jam hari ini", kita akan menggunakan NLP untuk mengenali "2 jam" dan mengubahnya menjadi nilai numerik 2 untuk digunakan dalam model.</p>

<h2>Langkah-langkah</h2>
<ol>
  <li><b>Pengumpulan Data:</b> Kumpulkan data berupa jumlah jam belajar dan nilai yang diperoleh oleh siswa.</li>

  <li><b>Pembentukan Model:</b> Gunakan algoritma Single Linear Regression untuk membangun model yang menghubungkan jumlah jam belajar (variabel independen) dengan nilai (variabel dependen).</li>
  <li><b>Evaluasi Model:</b> Uji model dengan data uji untuk melihat seberapa akurat model dalam memprediksi nilai.  Metrik evaluasi yang umum digunakan adalah R-squared, Mean Squared Error (MSE), dan Root Mean Squared Error (RMSE).</li>
  <li><b>Prediksi:</b> Setelah model divalidasi, gunakan model untuk memprediksi nilai siswa berdasarkan jumlah jam belajar yang baru.</li>
</ol>

### Contoh gambar grafik linear regression

```python



# Data contoh (jam belajar dan nilai)
jam_belajar = np.array([2, 3, 5, 7, 8])
nilai = np.array([60, 65, 75, 85, 90])

# Hitung koefisien regresi (contoh sederhana, biasanya dihitung dengan metode OLS)
a = 50  # Intercept (nilai Y ketika X=0)
b = 5   # Slope (perubahan Y untuk setiap perubahan satu unit X)


# Membuat garis regresi
garis_regresi = a + b * jam_belajar

# Membuat plot
plt.scatter(jam_belajar, nilai, label='Data Aktual', color='blue')
plt.plot(jam_belajar, garis_regresi, label='Garis Regresi', color='red')

# Menambahkan label dan judul
plt.xlabel('Jam Belajar')
plt.ylabel('Nilai')
plt.title('Prediksi Nilai Berdasarkan Jam Belajar')
plt.legend()
plt.grid(True)

# Menampilkan plot
plt.show()
```

## Train_Test_Split

```python



# Pisahkan data menjadi fitur (X) dan target (y)
X = df[['Hours']]
y = df['Scores']

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


```

### Visualiziong the data

```python
plt.figure(figsize=(10, 7))
# Plot training data in blue
plt.scatter(X_train, y_train, c="b", label="Training data")
# Plot testing data in green
plt.scatter(X_test, y_test, c="g", label="Testing data")
# Show the legend
plt.legend()
```

### Create a Model

```python
# 1. Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

# Fit the model
model.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100)
```

### Make some predictions

```python
# Make some predictions
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

### Make a function for Plotting

```python
# Let's create a plotting function
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=y_pred):
  """
  Plots training data, test data and compares predictions.
  """

  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c="b", label="Training data")
  # Plot testing data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing data")
  # Plot model's predictions in red
  plt.scatter(test_data, predictions, c="r", label="Predictions")
  # Show the legend
  plt.legend()
```

```python
# prompt: hitungkan mse dari model pakek tf.metrics

# Calculate MSE using tf.keras.metrics.MeanSquaredError
mse_metric = tf.keras.metrics.MeanSquaredError()
mse_metric.update_state(y_test, model.predict(X_test))
mse = mse_metric.result().numpy()

print(f"Mean Squared Error (MSE): {mse}")
```

```python
# prompt: make some function to reuse MAE and MSE

import tensorflow as tf

def calculate_metrics(model, X_test, y_test):
  """Calculates MAE and MSE for a given model and test data.

  Args:
    model: The trained TensorFlow model.
    X_test: The input features for the test set.
    y_test: The true labels for the test set.

  Returns:
    A dictionary containing the MAE and MSE values.
  """

  # Calculate MAE using tf.keras.metrics.MeanAbsoluteError
  mae_metric = tf.keras.metrics.MeanAbsoluteError()
  mae_metric.update_state(y_test, model.predict(X_test))
  mae = mae_metric.result().numpy()

  # Calculate MSE using tf.keras.metrics.MeanSquaredError
  mse_metric = tf.keras.metrics.MeanSquaredError()
  mse_metric.update_state(y_test, model.predict(X_test))
  mse = mse_metric.result().numpy()

  return {"mae": mae, "mse": mse}
```

```python

# Model 1: 1 layer, 100 epochs (same as original)
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model_1.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
model_1.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100, verbose=0) # Suppress output
y_pred_1 = model_1.predict(X_test)
print("Model 1 Evaluation:")
model_1.evaluate(X_test, y_test, verbose=0)

# Model 2: 2 layers, 100 epochs
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1)
])
model_2.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
model_2.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=100, verbose=0)
y_pred_2 = model_2.predict(X_test)
print("Model 2 Evaluation:")
model_2.evaluate(X_test, y_test, verbose=0)

# Model 3: 2 layers, 500 epochs
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1)
])
model_3.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
model_3.fit(tf.expand_dims(X_train, axis=-1), y_train, epochs=500, verbose=0)
y_pred_3 = model_3.predict(X_test)
print("Model 3 Evaluation:")
model_3.evaluate(X_test, y_test, verbose=0)
```
